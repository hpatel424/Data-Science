```python
#Q2.
class Vehicle:
  def __init__(self,name,baseFare):
    self.name=name
    self.baseFare = baseFare
  def fare(self):
    return self.baseFare

class Bus(Vehicle):
  def fare(self):
    return self.baseFare + 20

class Car(Vehicle):
  def fare(self):
    return self.baseFare + 50

class Train(Vehicle):
  def fare(self):
    return self.baseFare + 40

class Truck(Vehicle):
  def fare(self):
    return self.baseFare + 100

class Ship (Vehicle):
  def fare(self):
    return self.baseFare + 150

Bus1 = Bus('Bus',10)
Car1 = Car('Car',50)
Train1 = Train('Train',70)
Truck1 = Truck('Truck',100)
Ship1 = Ship('Ship',250)

vehiclesList = [Bus1, Car1, Train1, Truck1, Ship1]

TotalFare = 0
print("Calculating Fares:")

for i in vehiclesList:
    fare = i.fare()
    print(f"Fare for {i.name}: ${fare}")
    TotalFare += fare

print(f"Total Fare for all items: ${TotalFare}")
```

```python
#Q6
from abc import ABC, abstractmethod

class Shape(ABC):
  @abstractmethod
  def area(self):
    pass

class Circle(Shape):
  #PI = 3.14
  def __init__(self):
    self.PI = 3.14
    self.radius = 0

  def area(self,radius):
    return self.PI * radius * radius

  def circumference(self,radius):
    return 2 * self.PI * radius

class Rectangle(Shape):
  def __init__(self,length,width):
    self.length = length
    self.width = width

  def area(self):
    return self.length * self.width

c1 = Circle()
r1 = Rectangle(2,5)
print(f'Area of Cirle: {c1.area(2)}')
print(f'Circumference: {c1.circumference(4)}')
print(f'Area of Rectangle: {r1.area()}')
```

```python
#Q7
from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self._pi = 3.14          # protected member
        self.__radius = radius   # private member (name mangling)

    def area(self):
        # accessing private member inside the class
        return self._pi * self.__radius * self.__radius

    def circumference(self):
        return 2 * self._pi * self.__radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length      # public
        self._width = width       # protected

    def area(self):
        return self.length * self._width

shapes = [
    Circle(5),
    Rectangle(4, 6)
]

for s in shapes:
    print("Area:", s.area())

# Accessing private variable using name-mangling.
circle = Circle(3)
print("Access private radius (mangled name):", circle._Circle__radius)
print("Circumference:", circle.circumference())
```

```python
#Q9,Q10,Q11,Q12
class Triangle:
  number_of_sides = 3
  def __init__(self, angle1, angle2, angle3):
      self.angle1 = angle1
      self.angle2 = angle2
      self.angle3 = angle3

  def check_angles(self):
    sum = self.angle1 + self.angle2 + self.angle3
    if sum == 180:
      print('The sum is 180, angles forms triangle')
      return True
    else:
      print('The is not 180, does not form triangle')
      return False

  def acute(self):
    if(self.angle1 <90 and self.angle2 <90 and self.angle3<90):
      print('Acute Triangle!')
      return True
    else:
      print('NOT an acute triangle')
      return False

  def obtuse(self):
    if(self.angle1>90 or self.angle2>90 or self.angle3>90):
      print('Obtuse Triangle!')
      return True
    else:
      print('NOT an Obtuse triangle')
      return False

class Isosceles_triangle(Triangle):
  def isosceles(self):
    if self.check_angles() == True and (self.angle1 == self.angle2 or
                                      self.angle2 == self.angle3 or
                                      self.angle1 == self.angle3):
      print("Isosceles Triangle!")
      return True
    else:
      print("NOT an Isosceles Triangle")
      return False

class RightTriangle(Triangle):
    def right(self):
        if self.check_angles() and (self.angle1 == 90 or
                                    self.angle2 == 90 or
                                    self.angle3 == 90):
            print("This is a right triangle.")
            return True
        else:
            print("This is NOT a right triangle.")
            return False

class EquilateralTriangle(Triangle):
    def is_equilateral(self):
        if self.check_angles() and self.angle1 == self.angle2 == self.angle3 == 60:
            print("This is an equilateral triangle.")
            return True
        else:
            print("This is NOT an equilateral triangle.")
            return False

class IsoscelesRightTriangle(Isosceles_triangle, RightTriangle):
    def is_isosceles_right(self):
        # Check both properties using parent methods
        is_isosceles = self.isosceles()
        is_right = self.right()

        if is_isosceles and is_right:
            print("This is an Isosceles Right Triangle!")
            return True
        else:
            print("This is NOT an Isosceles Right Triangle.")
            return False

t1 = Triangle(60,75,45)
t1.check_angles()
t1.acute()
t1.obtuse()

t2=Isosceles_triangle(70,70,40)
t2.check_angles()
t2.isosceles()
print('-----------------------------------')
t = IsoscelesRightTriangle(45, 45, 90)
t.is_isosceles_right()
print('-----------------------------------')
# 60-60-60 triangle (not right)
t2 = IsoscelesRightTriangle(60, 60, 60)
t2.is_isosceles_right()
```
